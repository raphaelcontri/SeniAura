# Clustering Tests for Seinaura Capstone Project





The aim of this notebook is to test if we can apply clustering in our dashboard, see it here : https://seniaura.onrender.com/.



```python
import pandas as pd

df = pd.read_excel('/content/FINAL-DATASET-epci-11.xlsx')
print("DataFrame loaded successfully. First 5 rows:")
df.head()
```

```python
columns_to_drop = ['CODE_EPCI', 'LIBEPCI', 'nom_EPCI', 'Département_code', 'AIR01']
df_processed = df.drop(columns=columns_to_drop)

print(f"Original DataFrame shape: {df.shape}")
print(f"Processed DataFrame shape after dropping columns: {df_processed.shape}")
print("First 5 rows of df_processed after dropping columns:")
print(df_processed.head())
```

```python
print("Missing values per column in df_processed:")
print(df_processed.isnull().sum())
```

```python
numerical_cols = df_processed.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df_processed.select_dtypes(include=['object']).columns.tolist()

print(f"Identified {len(numerical_cols)} numerical columns:")
print(numerical_cols[:5], "...\n") # Print first 5 if many

print(f"Identified {len(categorical_cols)} categorical columns:")
print(categorical_cols)
```

```python
for col in numerical_cols:
    if df_processed[col].isnull().any():
        df_processed[col] = df_processed[col].fillna(df_processed[col].mean())

print("Missing values in numerical columns after imputation:")
print(df_processed[numerical_cols].isnull().sum().sum())
```

```python
for col in categorical_cols:
    if df_processed[col].isnull().any():
        df_processed[col] = df_processed[col].fillna(df_processed[col].mode()[0])

print("Missing values in categorical columns after imputation:")
print(df_processed[categorical_cols].isnull().sum().sum())
```

```python
for col in categorical_cols:
    if df_processed[col].isnull().any():
        # Convert the column to string type before calculating mode
        # This handles mixed types (like str and datetime) which can cause comparison errors
        # when mode() tries to sort values.
        df_processed[col] = df_processed[col].fillna(df_processed[col].astype(str).mode()[0])

print("Missing values in categorical columns after imputation:")
print(df_processed[categorical_cols].isnull().sum().sum())
```

```python
df_encoded = pd.get_dummies(df_processed, columns=categorical_cols, drop_first=False)

print(f"Shape of DataFrame after one-hot encoding: {df_encoded.shape}")
print("First 5 rows of df_encoded after one-hot encoding:")
print(df_encoded.head())
```
